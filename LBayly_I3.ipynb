```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your csv file
df = pd.read_csv('/content/Civilian_Complaint_Review_Board__Police_Officers_20260408.csv')

# Display the first 5 rows and column information to help identify column names
print("First 5 rows of the DataFrame:")
display(df.head())
print("\nDataFrame Info:")
df.info()

# Placeholder for your filtering logic:
#filtered_df = df[df['Total Complaints'] > 0] # Example condition 1
# To add a second condition, you can combine them like this:
filtered_df = df[(df['Total Complaints'] > 0) & (df['Active Per Last Reported Status'] == 'Yes')]

print("\nFiltered DataFrame Head:")
display(filtered_df.head())

# Select the top 25 values for the graph
data_for_graph = filtered_df['Current Command'].value_counts().head(25)

plt.figure(figsize=(12, 8)) # Increased figure size for better readability
sns.barplot(x=data_for_graph.index, y=data_for_graph.values, palette='viridis', hue=data_for_graph.index, legend=False)
plt.title(f'Bar Graph of Top 25 {data_for_graph.name} (Filtered Data)')
plt.xlabel(data_for_graph.name)
plt.ylabel('Count')
plt.xticks(rotation=90, ha='right') # Rotated labels for better visibility of many categories
plt.tight_layout()
plt.show()
```
