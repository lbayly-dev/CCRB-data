```python
# The pro of using this JSON API is that the data is automatically updated, meaning that the user does not need to go in and figure
# out whether this data is still relevant to use. The downside of this access route is that depending on the data, parsing a json can
# be complicated. JSON files are also not particularly easy for a human to read on their own, so having usable code to view and
# interpret it is important.

import requests

url = "https://api.worldbank.org/v2/country/US/indicator/SP.POP.TOTL?format=json"

response = requests.get(url)
data = response.json()

records = data[1]

for entry in records[:10]:
    year = entry["date"]
    value = entry["value"]
    print(f"{year}: {value}")
```

```python
# The benefit of this method is that it updates the output every time the code is executed, meaning that the user receives
# up to date information. The disadvantage is that it only shows the headlines of the articles, and does not provide a way
# for the user to see the entirety of the article.

import requests
from bs4 import BeautifulSoup

url = "https://www.bbc.com/news"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

headlines = soup.find_all("h2")

for h in headlines[:10]:
    print("-", h.get_text())
```
